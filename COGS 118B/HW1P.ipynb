2. Then find their covariance and assign it to `cov_xy_p1`<br>

The covariance matrix for two variables X and Y looks like: <br>
$$
\begin{pmatrix}
\text{cov}(X, X) & \text{cov}(X, Y) \\
\text{cov}(Y, X) & \text{cov}(Y, Y)
\end{pmatrix}
$$

Now that you have visualized the data, 
1. Find the means of each column X, Y and Z and assign them to `x_p2_mean`, `y_p2_mean` and `z_p2_mean` respectively.
2. Then find their covariance and assign it to `cov_xyz_p2`<br>

The covariance matrix for three variables X, Y and Z looks like: <br>
$$
\begin{pmatrix}
\text{cov}(X, X) & \text{cov}(X, Y) & \text{cov}(X, Z)\\
\text{cov}(Y, X) & \text{cov}(Y, Y) & \text{cov}(Y, Z)\\
\text{cov}(Z, X) & \text{cov}(Z, Y) & \text{cov}(Z, Z)
\end{pmatrix}
$$

```python
data = np.stack((x_p2, y_p2, z_p2))

x_p2_mean = np.mean(x_p2)
y_p2_mean = np.mean(y_p2)
z_p2_mean = np.mean(z_p2)

cov_xyz_p2 = np.cov(data)

print(f"Mean of x: {x_p2_mean}, Mean of y: {y_p2_mean}, Mean of z: {z_p2_mean}")
print(f"Covariance of (x, y, z):\n {cov_xyz_p2}")
```

Now, let's shift the gaussian to a new mean. <br>
1. We achieve this by adding the SHIFT vector directly to the data <br>
2. Let's use the SHIFT vector = [2, 3, 0], which will move the mean of the distribution to (2, 3, 0) <br>
3. Assign the shifted data to the variable `shifted_data` <br>
4. Plot the data again.

```python
shift_vector = np.array([2, 3, 0]).reshape(3, 1)
shifted_data = data + shift_vector


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs = shifted_data[0, :], ys = shifted_data[1, :], zs = shifted_data[2, :], c = shifted_data[2, :], cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("3-D Scatterplot of Shifted Data")
```

Let's try increasing the variance along the x-axis. <br>
1. We achieve this by multiplying the data with the covariance matrix: [[2, 0, 0], [0, 1, 0], [0, 0, 1]] <br>
2. This increases the spread (variance) along the X-axis while keeping Y and Z unchanged. <br>
3. Assign the scaled data to the variable `scaled_data` <br>
4. Plot the data again. Notice the wider scale along the X-axis

```python
scaling_matrix = np.array([
                           [2, 0, 0],
                           [0, 1, 0],
                           [0, 0, 1]
])

scaled_data = np.dot(scaling_matrix, data)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs = scaled_data[0, :], ys = scaled_data[1, :], zs = scaled_data[2, :], c = scaled_data[2, :], cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("3-D Scatterplot of Scaled Data")
```

```python
assert scaled_data[0,0] == -10.0
```

Let's introduce a correlation between the X and Y variables <br>
1. We achieve this by multiplying the data with the covariance matrix: [[1, 0.2, 0], [0.2, 1, 0], [0, 0, 1]] <br>
2. Assign the rotated data to the variable `rotated_data` <br>
3. This introduces a mild positive correlation between X and Y, resulting in an upward-tilted ellipse. <br>
4. Plot the data again

```python
correlation_matrix = np.array([
                        [1, 0.2, 0],
                        [0.2, 1, 0],
                        [0, 0, 1]
])
                                

rotated_data = np.dot(correlation_matrix, data)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs = rotated_data[0, :], ys = rotated_data[1, :], zs = rotated_data[2, :], c = rotated_data[2, :], cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("3-D Scatterplot of Rotated Data")
```
